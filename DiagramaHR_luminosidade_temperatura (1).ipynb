```python
# !pip install pandas
import pandas as pd
import matplotlib.pyplot as plt

# Abrindo o arquivo excel com pandas

tabela_estrelas = pd.read_excel('BHAC15 _tabela_completa.xlsx')
#print(tabela_estrelas)


lista_luminosidade = []

    # Criando uma variável para chamar a coluna da luminosidade

luminosidade = tabela_estrelas['L/Ls']
#print(luminosidade)

for i, v in enumerate(luminosidade):
    lista_luminosidade.append(v)
# print(lista_luminosidade)

# Agrupando os valores da temperatura uma lista

lista_temperatura = []
temperatura = tabela_estrelas['Teff']

for t, v in enumerate(temperatura):
    lista_temperatura.append(v)
#print(lista_temperatura)


x = lista_temperatura
y = lista_luminosidade
#print(x[::-1])
#print(y)

plt.plot(x,y, color = '#32d7af')
plt.xlabel('Log(Teff)[K]')
plt.ylabel('Log(L/Ls)')
plt.axis(xmin = 3.8, xmax =3.6, ymin = -0.4, ymax = 0.4)
plt.xticks([3.75, 3.70, 3.65, 3.6])
plt.yticks([-0.4, -0.2, 0.0, 0.2, 0.4])
plt.title('Diagrama HR')
plt.show()

```
